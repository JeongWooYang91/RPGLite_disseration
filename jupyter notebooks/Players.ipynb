```python
#Separate the json format 'count_character_has_won column'
wins = pd.json_normalize(players2['count_character_has_won'])
wins = wins.rename(columns={'K': 'Knight_win', 'A': 'Archer_win','R': 'Rogue_win','B':'Barbarian_win', 'H': 'Healer_win','W': 'Wizard_win','M': 'Monk_win','G': 'Gunner_win'})
wins.head(10)
```
